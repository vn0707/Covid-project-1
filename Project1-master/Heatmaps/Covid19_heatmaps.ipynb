```python
# Import Dependencies
import pandas as pd
import gmaps
from config import gkey

gmaps.configure(api_key=gkey)
```

```python
# Create a DataFrame of frames using a dictionary of lists
#file_path = "time_series_covid19_confirmed_US.csv"
file_path = "time_series_covid19_deaths_US.csv"
covid_data_df = pd.read_csv(file_path)
covid_data_df
```

```python
covid_data_df = covid_data_df.dropna()
```

```python
cleaned_covid_data_df = covid_data_df.loc[covid_data_df["Lat"]!=0.0,]
cleaned_covid_data_df
```

```python
cleaned_covid_data_df.columns
```

```python
#cleaned_covid_data_df.info()
```

```python
locations = cleaned_covid_data_df[["Lat", "Long_"]]
confirmed_case = cleaned_covid_data_df["4/18/20"]

fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=confirmed_case, 
                                 dissipating=False,  max_intensity=1000,
                                 point_radius=2)

fig.add_layer(heat_layer)

fig
```
